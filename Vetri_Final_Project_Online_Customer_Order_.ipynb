## Stage 2 – Data Cleaning and Pre-processing

Handle missing values (impute or drop)

Handle duplicates

Treat outliers if required

Check skewness and apply transformations

Convert data types if needed

Feature transformations (date parts, derived fields if required for analysis)

```python
# Handling missing values
df.isnull().sum()
```

```python
#replace nulls in review_score column with mean
df["review_score"].fillna(df["review_score"].mean(), inplace=True)
print(df[['review_score']].head(10))
```

```python
#replace nulls in gender column with unknown
df["gender"].fillna("Unknown", inplace=True)
print(df["gender"].value_counts(dropna=False))
print(df[['gender']].head(10))
```

```python
# Removing duplicates
print("\nShape before removing duplicates:", df.shape)
df = df.drop_duplicates()
print("Shape after removing duplicates:", df.shape)
```

```python
# Correcting data types
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
print("after correction data types", df.dtypes)
```

```python
# Creating derived columns
df['total_sales']=(df['quantity'] * df['price'])
df
```

```python
#Filtering data
filtered_df = df[(df['quantity'] > 3) & (df['payment_method'] == 'Credit Card')]
```

```python
#Filtering and aggregatd
aggregated_sales = df.groupby('category_name')['total_sales'].sum().reset_index()
# Display results
print("Filtered Data (first 5 rows):")
print(filtered_df.head())

print("\nAggregated Total Sales per Category:")
print(aggregated_sales)
```

```python
#histogram numerical column
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(7,5))
sns.histplot(df['total_sales'],bins=30,kde=True,color="green")
plt.title("delivery of total sales")
plt.show()
```

## Interpretation of your above chart

1.The histogram displays how product prices are distributed across all customer orders.

2.Most products are priced within the low to mid-range, indicating customers prefer affordable items. Very few high-priced products exist.

3.Used feature:product_price.

4.It shows the pricing pattern and helps identify the common price range where maximum sales occur.

```python
# Box Plot a numerical column(age)
plt.figure(figsize=(6,5))
sns.boxplot(x=df['age'],color="red")
plt.title(" age ")
plt.show()
```

## Interpretation of your above chart

1.The median age lies between 35–40 years, showing that middle-aged customers dominate the online sales.


2.There are some outliers representing younger and older customers, but the majority fall within the 25–50 age group


```python
# Count plot for categorical variable
sns.countplot(data=df, x="payment_method",palette="viridis")
plt.title("Count Plot of Payment Method")
plt.show()
```

## Interpretation of your above chart

1.Credit Card and Debit Card are the most common methods, showing customer preference for digital transactions.


2.Fewer transactions occur via Cash on Delivery, suggesting trust in online payments.

```python
# ---------- Bivariate Analysis ----------

# Scatter plot
sns.scatterplot(data=df, x="age", y="price", hue="gender")
plt.title("Scatter Plot: Age vs Price")

plt.tight_layout()
plt.show()
```

## Interpretation of your above chart

1.Both male and female customers show similar purchase patterns.


2.Higher-priced products are typically bought by middle-aged customers (30–45 years).


3.Younger customers tend to prefer lower-priced items.


```python
# Bar plot (avg price by category)
sns.barplot(data=df, x="category_name", y="price")
plt.title("Bar Plot: Avg Price by Category")
plt.show()
```

## Interpretation of your above chart

1.Electronics and Sports & Outdoors categories have the highest average prices.


2.Books & Stationery and Fashion are lower-priced categories.

```python
# Correlation heatmap
corr = df[["quantity", "price", "review_score", "age"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

## Interpretation of your above chart
1.Low correlation between quantity and price, meaning expensive products don’t necessarily sell in larger quantities.


2.Slight positive correlation between price and review_score, suggesting premium products receive better reviews.


3.Age shows weak correlation with all numeric variables, indicating diverse purchasing behavior across age groups.


```python
# ---------- Multivariate Analysis ----------

# Pair plot of multiple variables with hue
sns.pairplot(df[["quantity", "price", "review_score", "age", "gender"]], hue="gender")
plt.suptitle("Pair Plot of Multiple Variables", y=1.02)
plt.show()
```

## Interpretation of your above chart

1.Provides a holistic view of how these variables interact.


2.Reveals clusters among genders and highlights data patterns visually
